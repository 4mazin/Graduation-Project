```python
import pandas as pd
import spacy
import numpy as np
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df1 = pd.read_csv("Metadata_Dataset/all_courses.csv", usecols=['course_name', 'description', 'url'])
df1.head()
```

```python
df2 = pd.read_csv("Metadata_Dataset/Online_Courses.csv", usecols=['Title', 'Short Intro', 'URL'])
df2
```

```python
df3 = pd.read_csv("Metadata_Dataset/Udacity_final.csv", encoding = "cp1252", usecols=['Course_Name', 'About', 'Skills', 'Link'])
df3
```

```python
rename_map_df3 = {
    "Course_Name": "course_name",
    "About": "description",
    "Link" : "url"
}

df3 = df3.rename(columns = rename_map_df3)
df3.head()
```

```python
df3['description'] = df3['description'] + ' ' + df3['Skills']
```

```python
rename_map_df2 = {
    "Title" : "course_name",
    "URL" : "url",
    "Short Intro" : "description"
}

df2 = df2.rename(columns = rename_map_df2)
df2.head()
```

```python
metadata = pd.concat([df1, df2, df3], ignore_index = True)
```

```python
metadata.info()
```

```python
metadata = metadata.dropna(subset = ['description'])
metadata = metadata.reset_index(drop = True)
metadata.info()
```

```python
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load("en_core_web_sm")

def preprocess(text):
    doc = nlp(text, disable=['ner', 'pos'])
    lemmas = [token.lemma_.lower() for token in doc]
    clean_lemmas = [
        lemma for lemma in lemmas
        if lemma.isalpha() and lemma not in STOP_WORDS and lemma != '-PRON-'
    ]
    return ' '.join(clean_lemmas)
```

```python
metadata['clean_text'] = metadata['description'].apply(preprocess)
metadata['clean_text']
```

```python
clean_courses = metadata[['course_name', 'description', 'url', 'clean_text']]
clean_courses.info()
```

```python
clean_courses = clean_courses.drop_duplicates()
```

```python
clean_courses.info()
```

```python
clean_courses = clean_courses.dropna(subset=['clean_text'])
clean_courses = clean_courses.reset_index(drop=True)
clean_courses.info()
```

```python
clean_courses = clean_courses.drop_duplicates(subset=['course_name'], keep='first')
clean_courses.info()
```

```python
clean_courses = clean_courses.drop_duplicates(subset=['description'])
clean_courses.info()
```

```python
clean_courses.to_csv("clean_courses.csv")
```
